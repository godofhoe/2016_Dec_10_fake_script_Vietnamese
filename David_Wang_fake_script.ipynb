```python
def read_script(inp_file_name, sp = " "):
    """Read a fake script and return a list contains all the characters in the script.
    """
    text =  open(inp_file_name,"r")
    string_list = []

    for line in text:
        l = line.split()
        for word in l:
            for char in list(word):
                string_list.append(char)

    text.close()        
    return string_list
```

```python
def generate_DDWang_text(inp_file_name,out_file_name,word_len):
    """a function which reads a file, generateing Vietnamese script and saving it as a file with name file_name.

    Parameters
    ---
    file_name: str
        the name of a file containing novel texts
    
    word_len: int
        an interger >= 0
        
    Returns
    ---
    None
    
    Raises
    ---
    AssertionError

    """
    assert word_len >= 0
    f2 = open(out_file_name , 'w')
    string_list = read_script(inp_file_name, sp = " ")
    random.shuffle(string_list)
    list_2 = []
    tmp = ''
    for e in string_list:
        tmp = tmp + e
        if len(tmp) == word_len:
            list_2.append(tmp)
            tmp = ''
   
    word_count = 0
    for item in list_2:
        if word_count < 20:
            f2.write(item+" ")
            word_count += 1
        else:
            word_count = 0
            f2.write(item+"\n")
    f2.close()
    print("A DDWang fake script is successfully created!")
    return None
```
